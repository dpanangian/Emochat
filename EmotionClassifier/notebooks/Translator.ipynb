```python
from datasets import load_dataset
#from labels import mapping # listing the original class names

import pandas as pd
import numpy as np

from tqdm.notebook import tqdm
from matplotlib import pyplot as plt

from torch.utils.data import DataLoader
```

```python
go_emotions = load_dataset("go_emotions")
data = go_emotions.data

train = go_emotions.data["train"].to_pandas()
valid = go_emotions.data["validation"].to_pandas()
test = go_emotions.data["test"].to_pandas()

print(train.shape, valid.shape, test.shape)
```

```python
print(train)
```

```python
test = train.head(1)
german_emotions = train.copy(deep=True)
print(german_emotions)
german_top = german_emotions.head(10)
```

```python
from googletrans import Translator

translator = Translator()
translations = translator.translate(test["text"], dest="de")
print(translations)
for translation in translations:
  print(translation.origin, ' -> ', translation.text)
```

```python
import requests

def translate(text):
  params = {'auth_key': '8d3b7156-cbdc-5732-eb9b-7152d6f4b15e:fx', 'text': text, 'target_lang': 'DE'}
  response = requests.get("https://api-free.deepl.com/v2/translate", params=params)
  if response.status_code == 200:
    return response.json()["translations"][0]["text"]
  else:
    return text

print(german_emotions.head(1))
for val in german_emotions["text"]:
  print("English: " + val)
  german = translate(val)
  german_emotions["text"] = german_emotions["text"].replace(val, german)
  print("German: " + german)

print(german_emotions)
```
